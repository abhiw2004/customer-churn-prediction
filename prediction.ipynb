```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
##LOad the trained ANN model, scaler and encoders
model = load_model('model.h5')


##Load the encoders and scalers
with open('onehot_encoder_geo.pkl','rb') as file: 
    onehot_encoder_geo = pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    "CreditScore" : 600,
    "Geography" : "France",
    "Gender" : "Male",
    "Age" : 40,
    "Tenure" : 3,
    "Balance" : 60000,
    "NumOfProducts" : 2,
    "HasCrCard" : 1,
    "IsActiveMember" : 1,
    "EstimatedSalary" : 50000
} 
```

```python
##converting into a dataframe
input_data = pd.DataFrame([input_data])
```

```python
##Onehot encode Geography 
geo_encoded = onehot_encoder_geo.transform(input_data[['Geography']])
geo_encoded_df = pd.DataFrame(geo_encoded, columns = onehot_encoder_geo.get_feature_names_out(['Geography']))
input_data = input_data.drop('Geography',axis = 1)
input_data = pd.concat([input_data,geo_encoded_df], axis = 1)
```

```python
##Label encoding Gender feature
input_data['Gender'] = label_encoder_gender.transform(input_data['Gender'])
```

```python
# Scale
input_data_scaled = scaler.transform(input_data)
```

```python
# Predict
prediction = model.predict(input_data_scaled)
predicted_probab = prediction[0][0]
```

```python
if predicted_probab > 0.5:
    print("Customer is likely to churn")
else:
    print("Customer in not likely to churn")
```
